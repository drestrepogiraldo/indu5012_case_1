## An Overview of Predictive Analytics

In contrast to descriptive analytics, </b>predictive analytics</b> seek to forecast the likely future state of the world through a deeper understanding of the relationships between data inputs and outcomes. So there is much more that can go wrong. Once you have a good descriptive model, you cannot use it to make good forecasts and that is  as some apparent relationships may not be statistically significant, as </b>correlation does not imply causation</b>. As a result, when the objective is to forecast the future, the use of more sophisticated modeling approaches and more rigorous validation procedures is fundamental generate forecasts that are coherent with reality.

As an everyday example of the difference between descriptive and predictive analytics, consider television programs that cover the stock market. Every day, TV hosts explain why the stock market behaved the way it did the previous day. However, none of them can accurately forecast what the market will do tomorrow. </b>Hindsight may be 20–20, but foresight is not</b>.

</b>Data Mining and Pattern Recognition</b>, the starting point for predictive analytics, is often processing data to identify meaningful relationships and patterns between different features of interest. For example:
- Clustering analysis might be used to segment customer populations into groups that go beyond simple demographic or psychographic characteristics. 
- Machine learning techniques can be used to identify objects and trajectories for autonomous vehicle scene recognition and navigation. 

Machine learning and artificial intelligence techniques can arguably answer a broader set of questions (e.g., image recognition), but trade the transparent simplicity and interpretability of classical models for a harder-to-explain </b>“black box”</b> capable of representing more complex relationships. 

The tools used in machine learning can be broadly separated into <b>supervised and unsupervised learning</b>:
- **Supervised statistical** learning refers to the use of statistical models to predict an output based on one or more inputs. 
- **Unsupervised statistical** learning involves inputs, but no known outputs, that try to find relationships and structures within the inputs. 

Within the problems that can be treated with supervised statistical learning, there are 
- **Classification problems**, that focus on classifying populations into distinct groups (e.g., classifying a bank transaction as fraudulent or non-fraudulent whenever it is made)
- **Regression problems**, that focus on estimating the impact of a set of input variables on a metric of interest (e.g., estimating someone's future salary based on their characteristics).

## Beware!
### Causation vs Correlation

It is important to remember the difference between causation and correlation; while **correlation** is a statistical measure that represents the magnitude and the direction of the relationship between two features, **causation** implies that one event is caused by another. Keeping this in mind, it is worth noting that on most of the studies where statistical relationships between features are studied, only correlation can be determined; in order to establish causation, experiments under controlled conditions must be undertaken (e.g. medical trials). However, not distinguishing the difference between correlation and causation is a common mistake by unexperienced data scientists, and this can lead to the proposal of some proposal of some bold statements.

For example, Rich Caruana, a researcher at Microsoft Research Redmond presented a case in which a predictive machine learning model was trained to predict risk of death in pneumonia patients ([Caruana et al. , 2015](https://www.scinapse.io/papers/1996796871)). The model ended up learning that, according to the sample which it was given, patients who also had asthma as comorbid condition had a better probability of survival. <b> What do you think this could mean? How can this happen? </b>

As ilogical as the relationship may sound, the model did not make an error; asthma was indeed predictive of survival, this was a true association in the training data. However, as you may have thought, the relationship between asthma and death by pneumonia is not causal. The asthma patients were more likely to survive because their treatment was more aggressive (they were automatically admitted directly into the intensive care units). 

Now, think about the repercussions that the acceptance of the statement "people that have asthma are more likely to survive pneumonia than a person that does not have asthma" could have. Therefore, be wary about the difference between correlation and causation, and also be mindful about the fact that only correlation between variables can be identified on the most of the statistical analysis that are made, as there could be complex dynamics (which are unidentifiable with the features that are measured) that create the correlation between variables. 



### Spurious correlations

"Spurious correlation" is another concept that you have to keep in mind when analyzing relationships between variables. A spurious correlation is a connection between two variables that may seem causal, but it is not. For example, do you think that eating cheese contributes in some way to dying by becoming tangled in your bedsheets? Well, you might want to think again after you see what is shown in [this](https://www.tylervigen.com/spurious-correlations) web page.  

As these examples show, it is important to always be aware of the fact that not all apparent correlations can be atributed to a dynamic that relates the two variables. The relationship between two variables that seem correlated but are actually not can be atributed to unknown common factors that affect both variables or only to pure randomness. In order to mitigate the appearance of spurious correlations in analysis made, it is important to make sure that the sample that is being analyzed is representative of the population and that it has a decent size. Also, it is worth noting that to determine whether a correlation is significant null hypothesis tests have to be done (more about this wil be shown in the course).  

## Supervised and unsupervised learning

In supervised learning, for each observation of the predictor measurement(s) $\textbf{x}_i, i = 1,...,n$ there is an associated response measurement $y_i$. Then, the intent is to fit a model that relates the response to the predictors, with the aim of accurately predicting the response for future observations (prediction) or better understanding the relationship between the response and the predictors (inference).The majority of the examples that are going to be treated in this course will be ones of supervised learning.

In contrast, unsupervised learning describes a more challenging situation in which for every observation $i = 1,...,n$, a vector of measurements $\textbf{x}_i$ is observed, but there is no associated response $y_i$. In unsupervised learning problems it is not possible to fit a linear regression model, since there is no response variable to predict. A situation of this type is referred to as unsupervised because there is no response variable that can help supervise our analysis. Now, if there is no response variables available to develop a classification or regression problem with, what kind of analysis can there be done? There can be an analysis of the relationship between the observed measurments $\textbf{x}_i$. One tool of unsupervised learning is the cluster analysis, or clustering. In clustering, the main goal is to group the observations in different groups (clusters) based on the input features that were gathered in each of those observations. For example, in a market segmentation analysis, it can be worthwhile to try to find different groups of potential costumers based on characteristics such as age, family income, location and spending habits. By grouping your costumers this way, you may gain some insight about the different types of people that attend your bussiness that may have different general interests and needs that can be addressed by products and/or services that you may provide.

#### Relationship between input and output variables

When treating a general problem of supervised learning, there are output variables (or dependent variables) and input variables (or independent variables). Output variables are the variables that should be predicted, which are going to be noted in this course as $\textbf{Y}$, where $\textbf{Y}$ is a vector containing all of the value for the output variables of the observations and $y_i$ is the value of the output variable for the $i^{th} $ observation. Input variables are the variables that are used as information to make the prediction of the output variables, which are going to be noted in this course as $\textbf{x}$, where $\textbf{x}$ is a matrix containing all of the values of the input variables for each one of the observations and $x_{ij}$ is the value of the $j^{th}$ input feature of the $i^{th} $ observation. On that account, the relationship between the input variables and the output variables can be expressed as: <center>$\textbf{Y} = \begin{bmatrix}
           y_{1} \\
           y_{2} \\
           \vdots \\
           y_{n}
         \end{bmatrix}$ </center> <center>$\textbf{x} = \begin{bmatrix}
           x_{11} & x_{12} & x_{13} & \dots & x_{1m}\\ 
           x_{21} & x_{22} & x_{23} & \dots & x_{2m}\\
           \vdots \\
           x_{n1} & x_{n2} & x_{n3} & \dots & x_{nm}\\
         \end{bmatrix}$ </center> <center>$\textbf{Y}= f(\textbf{x}) + \epsilon$</center>

#### Reducible vs irreducible error

When seeing machine learning like this, it is clear that the intent in machine learning is to estimate the function $f$ to the best of our abilities. Nevertheless, it is worth noting that when determining the relationships between variables there is a reducible error and an irreducible error when doing its estimation. The reducible error refers to the error that can be reduced by making a better estimation (i.e. by gathering more information, using new features to make the estimation, using models that fit the data better) and the irreducible error refers to an error that cannot be reduced even if the estimation is perfected. This concept of reducible and irreducible error can be seen in the definition that was used to relate the output variable with the input variables as the expected value of the squared difference between the estimation and a new observation:  <center>$E(\textbf{Y}-\hat\textbf{Y})^2= E(f(\textbf{x}) + \epsilon - \hat{f}(\textbf{x}))^2=(f(\textbf{x}) - \hat{f}(\textbf{x}))^2+ var(\epsilon)$</center> 

where $(f(\textbf{x}) - \hat{f}(\textbf{x}))^2$ is the term that refers to the reducible error, as $\hat{f}(\textbf{x})$ can be closer to $f(\textbf{x})$ as the estimation is better made, and $var(\epsilon)$ refers to the irreducible error, as this has nothing to do with the estimation made. The importance of the distinguishment between reducible and irreducible error is the fact that when the effort to reduce the squared error to zero (i.e. reducing both the reducible and irreducible error to zero) is made, one can be <b>overfitting</b> the data when estimating $\hat{f}(\textbf{x})$. However, when a poor effort is made to reduce the squared error one might <b>underfit</b> the data when estimating $\hat{f}(\textbf{x})$. 

#### How to estimate $f(\textbf{x})$

In order to estimate the relationship between input and output variables, one must first obtain a sample of the population that is going to be used to determine these relationships, these observations are called the <b>training set</b>, as these observations are used in order to train the statistical models that are going to be used to relate input and output variables. For example, in the previous case where the data of 30 people was taken, these 30 observations were the training set for that specific analysis.

Then, it is necessary to pre-determine the way in which the input and output variables relate as in this course only parametric models will be treated. Parametric models are models in which one assumes the functional form of $f$. For example, one simple assumption is that $f$ is a linear function as follows:<center>$f(\textbf{x})=\beta_0 + \beta_1X_1+\beta_2X_2+\dots+\beta_nX_n$</center> 

The asssumption about $f$ that is made in parametric models makes the estimation easier as only a few parameters have to be estimated. However, as an assumption of the function $f$ is being made, its estimation will be worse as the real $f$ has a form that differs more from the one being assumed.To avert this problem, one could use a form of $f$ that can take different shapes, depending on the values that some parameters have. Nonetheless, by having a more complicated $f$, its estimation will be slower and the number of observations required in the training set to fit the model will be higher than when compared to a simpler $f$ due to the fact that the more complicated $f$ will have more parameters that have to be estimated. Also, as models turn more complicated, its interpretation becomes a more difficult task.

Now, non-parametric models do not make an assumption about the functional form of $f$, so a better approximation of the function $f$ can be made. Nonetheless, as these model try to estimate $f$ without making any assumption, they require a very big number of observations iun order to make a correct estimation of $f$.


## Assesing model accuracy

### Error metrics for regression problems
In order to know if how well a machine learning model performs, there are different error metrics that can be used, depending on the intent of the statistical analysis that is being made. The most common error metric used in regression problems is the <b>mean squared error (MSE)</b>, which is calculated based on the values of the output variables ($y_i$) and the estimations made by the model ($\hat{f}(\textbf{x}_i)$) as follows:<center>$MSE=\frac{1}{n}\sum_{i=0}^m{(y_i-\hat{f}(\textbf{x}_i))^2}$</center>

It is noticeable that when the predictions made by the model are close to the value of the output variables, the MSE will have a value close to zero. In the same way, as the preddictions deviate farther from the value of the output variables, the value of the MSE will be higher. Therefore, a model that has a lower MSE is preferred as its estimations are closer to the real values. Now, as this metric results as the mean of the squared difference between the prediction and the real value, this metric's units will be the squared of the units that are used to measure the value of the output variable (e.g. if we are trying to predict the age of someone in years, the MSE will be in years^2). On that account, there is the <b>root mean squared error (RMSE)</b>, which is just the squared root of the MSE, which is expressed in the same units as the ones used to measure the value of the output variables: <center>$RMSE=\sqrt{\frac{1}{n}\sum_{i=0}^m{(y_i-\hat{f}(\textbf{x}_i))^2}}$</center>

A third measure of error used for regression problems is the <b>mean absolute error (MAE)</b> which is not calculated with the squared difference between prediction and real vlaue, but the absolute difference between them: <center>$MAE=\frac{1}{n}\sum_{i=0}^m{|y_i-\hat{f}(\textbf{x}_i)|}$</center>

Finally, there is the <b>mean absolute percentage error (MAPE)</b>, which does not calculate simply the difference between prediction and real values, but shows, on average, how far the estimattion is from the real value, relative to the real value of the output variable:<center>$MAPE=\frac{1}{n}\sum_{i=0}^m{\frac{|y_i-\hat{f}(\textbf{x}_i)|}{y_i}}$</center>




### Training set, validation set and test set

The concept of training set has been introduced already; however, as models can get really complex, the models used today can exactly fit all of the observations in the training set, achieving  a score of 0 in everyone of the metrics previously stated. <b> So, one should always use the more complex models that fit the training data perfectly, right? </b> 

If that was so, we would be able to perfectly predict a great number of variables of interest. However, let us keep in mind that the purpose of machine learning models is to predict the values of variables of interest that come from an observation that was never seen before by the model and also to identify relationships between variables that are present in a population. For example, when constructing a model that helps to predict the probability of some having a particualr type of cancer it is obvious that what is needed is for the model to accurately predict the probability of new patients having cancer and not for the model to predict accurately the probability of having cancer for the patients used in the training set, which would not be useful because all of the values of the variables of the training set have to be known (i.e. we should know already which of the patients in the training data have cancer). Also, let us remember what was talked about when differentiating between reducible error and irreducible error: it is not desirable to try to reduce the error to zero, as one can be making the model adjust for random uncontrolled variables by introducing a big number of parameters that will make the model overfit the training data and perform poorly on new unseen data.  

On that account, the data collected for the creation of a statistical model should be divided into a <b>training set</b> and a <b>validation set</b>. The training set, as explained before, is the set of data that is used for the model estimation. The validation set is a set of data not shown to the model that is used as unseen observations used to asses the accuracy of the model when predicting values for new observations. When calculating the MSE on the training set, that metric is called training MSE. Likewise, the validation MSE is the MSE calculated over the validation set. Therefore, validation metrics have to be the ones analyzed when selecting the best model used to predict certain value (either in regression problems or in classification problems).

Lastly,the <b> test set </b> is a set of the data different from the training set and the validation set that is used to present the final accuracy results of a machine learning model that are unbiased. The need for a third set of data that is held until the final model is selected is the fact that the training set and the validation set is somehow seen by the models that are being fitted, which makes that the accuracy metrics calculated over these sets have some bias in them. It is already clear why the training metrics have some bias in them; however, there is also a bias in the validation metrics as the validation set is used to modify and select models, which means that the resulting models will have some type of bias towards the validation set, even if this set is not used to fit the model.
 

### Bias and variance trade-off

When assesing the accuracy of machine learning models, there will be a u-shape behaviour when graphing validation MSE vs model complexity (we will see an example of this behaviour next). This behaviour is caused because when the model is too simple it will underfit the data. Then, as the model complexity increases it can fit the data better so the validation MSE will be decreasing. However, as the model complexity continues increasing, the model will overfit the training data and will start to have a poorer behaviour on the validation MSE.

Now, this convex behaviour in the validation MSE vs model complexity is the result of two properties of machine learning methods that are competing: the variance and the bias of the model estimation. These properties affect the validation MSE as the expected value of the validation MSE for a specific observation $x_0$ can be expressed as following: <center> $E(y_0-\hat{f}(x_0))^2 = Var(\hat{f}(x_0))+[Bias(\hat{f}(x_0))]^2+Var(\epsilon)$ </center>

Therefore, as the variance or the bias of the estimation increases, the model will have a worse performance in the validation MSE. Also, it is worth noting that the term $Var(\hat{f}(x_0)) + [Bias(\hat{f}(x_0))]^2$ can never be negative, so the expected MSE will never be lower than $Var(\epsilon)$, which is the irreducible error that was previously mentioned.

Therefore, if a method has high variance then small changes in the training data can result in large changes in $\hat{f}$. On the other hand, the bias of the estimation refers to how deviated the model's estimation is from reality, mainly due to the simplification of real-life interactions/dynamics that are made by the assumptions used in the model. As more flexible models can capture the interactions between variables better, more complex models tend to have less bias. 

On that account, as a general rule, as the methods used become more complex (flexible), the variance will increase and the bias will decrease. The relative rate of change of these two quantities as the complexity increases determines whether the validation MSE increases or decreases. As the flexibility of a class of methods increases, the squared bias tends to initially decrease faster than the variance increases. Accordingly, the expected validation MSE declines. However, at some point increasing flexibility has a small impact on the bias, but it starts to significantly increase the variance. When this happens, the test MSE increases as the complexity of the model keeps increasing.
